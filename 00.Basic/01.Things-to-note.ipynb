# Convert Integer to Binary - Python Notes

## Problem
Convert an integer to its binary string representation.

**Example:** 7 → `"111"` or `"0b111"`

---

## Methods

### 1. Built-in `bin()` (Easiest)

```python
n = 7
print(bin(n))        # Output: '0b111'
print(bin(n)[2:])    # Output: '111' (remove '0b' prefix)

# Negative numbers
print(bin(-7))       # Output: '-0b111'
```

**Key Points:**
- Returns string with `'0b'` prefix
- Use `[2:]` to remove prefix
- Works for negative numbers (adds `-` sign)

---

### 2. Manual Conversion (Interview Style)

```python
def int_to_binary(n):
    if n == 0:
        return "0"
    
    binary = ""
    while n > 0:
        binary = str(n % 2) + binary  # prepend remainder
        n //= 2                       # divide by 2
    
    return binary

print(int_to_binary(7))   # Output: '111'
print(int_to_binary(10))  # Output: '1010'
```

**How it works:**
```
7 % 2 = 1, 7 // 2 = 3  →  "1"
3 % 2 = 1, 3 // 2 = 1  →  "11"
1 % 2 = 1, 1 // 2 = 0  →  "111"
```

---

### 3. Using `format()` (Alternative)

```python
n = 7
print(format(n, 'b'))     # Output: '111'
print(f'{n:b}')           # Output: '111' (f-string style)
```

---

### 4. Fixed Width Binary (Padding with 0s)

```python
n = 7

# 8-bit representation
print(bin(n)[2:].zfill(8))    # Output: '00000111'
print(format(n, '08b'))       # Output: '00000111'
print(f'{n:08b}')             # Output: '00000111'
```

---

### 5. Binary to Integer (Reverse)

```python
binary_str = "111"

# Method 1: int() with base 2
print(int(binary_str, 2))     # Output: 7

# Method 2: with '0b' prefix
print(int('0b111', 2))        # Output: 7
```

## LeetCode Common Patterns

### Pattern 1: Process Each Bit
```python
def process_bits(n):
    result = []
    while n:
        result.append(n & 1)  # get last bit
        n >>= 1               # right shift
    return result[::-1]       # reverse to get correct order

print(process_bits(7))  # Output: [1, 1, 1]
```

### Pattern 2: Check Specific Bit Position
```python
def check_bit(n, pos):
    return (n >> pos) & 1

# Check 2nd bit (0-indexed) of 7 (111)
print(check_bit(7, 2))  # Output: 1
print(check_bit(7, 3))  # Output: 0
```

### Pattern 3: Get Binary Length
```python
n = 7
length = len(bin(n)) - 2      # Method 1
length = n.bit_length()        # Method 2 (faster)

print(length)  # Output: 3
```
