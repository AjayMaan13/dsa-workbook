# Convert Integer to Binary - Python Notes

## Problem
Convert an integer to its binary string representation.

**Example:** 7 → `"111"` or `"0b111"`

---

## Methods

### 1. Built-in `bin()` (Easiest)

```python
n = 7
print(bin(n))        # Output: '0b111'
print(bin(n)[2:])    # Output: '111' (remove '0b' prefix)

# Negative numbers
print(bin(-7))       # Output: '-0b111'
```

**Key Points:**
- Returns string with `'0b'` prefix
- Use `[2:]` to remove prefix
- Works for negative numbers (adds `-` sign)

---

### 2. Manual Conversion (Interview Style)

```python
def int_to_binary(n):
    if n == 0:
        return "0"
    
    binary = ""
    while n > 0:
        binary = str(n % 2) + binary  # prepend remainder
        n //= 2                       # divide by 2
    
    return binary

print(int_to_binary(7))   # Output: '111'
print(int_to_binary(10))  # Output: '1010'
```

**How it works:**
```
7 % 2 = 1, 7 // 2 = 3  →  "1"
3 % 2 = 1, 3 // 2 = 1  →  "11"
1 % 2 = 1, 1 // 2 = 0  →  "111"
```

---

### 3. Using `format()` (Alternative)

```python
n = 7
print(format(n, 'b'))     # Output: '111'
print(f'{n:b}')           # Output: '111' (f-string style)
```

---

### 4. Fixed Width Binary (Padding with 0s)

```python
n = 7

# 8-bit representation
print(bin(n)[2:].zfill(8))    # Output: '00000111'
print(format(n, '08b'))       # Output: '00000111'
print(f'{n:08b}')             # Output: '00000111'
```

---

### 5. Binary to Integer (Reverse)

```python
binary_str = "111"

# Method 1: int() with base 2
print(int(binary_str, 2))     # Output: 7

# Method 2: with '0b' prefix
print(int('0b111', 2))        # Output: 7
```

---

## LeetCode Common Patterns

### Pattern 1: Process Each Bit
```python
def process_bits(n):
    result = []
    while n:
        result.append(n & 1)  # get last bit
        n >>= 1               # right shift
    return result[::-1]       # reverse to get correct order

print(process_bits(7))  # Output: [1, 1, 1]
```

### Pattern 2: Check Specific Bit Position
```python
def check_bit(n, pos):
    return (n >> pos) & 1

# Check 2nd bit (0-indexed) of 7 (111)
print(check_bit(7, 2))  # Output: 1
print(check_bit(7, 3))  # Output: 0
```

### Pattern 3: Get Binary Length
```python
n = 7
length = len(bin(n)) - 2      # Method 1
length = n.bit_length()        # Method 2 (faster)

print(length)  # Output: 3
```

# Count Digits in a Number - Quick Notes

## Problem
Count how many digits are in a number.

**Example:** 12345 → 5 digits

---

## Method 1: Loop Method (Normal/Brute Force)

### Python
```python
def count_digits(n):
    count = 0
    while n > 0:
        count += 1
        n //= 10  # remove last digit
    return count

print(count_digits(12345))  # Output: 5
```

### C++
```cpp
int countDigits(int n) {
    int count = 0;
    while (n > 0) {
        count++;
        n /= 10;  // remove last digit
    }
    return count;
}
```

**How it works:**
```
12345 → remove 5 → count = 1
1234  → remove 4 → count = 2
123   → remove 3 → count = 3
12    → remove 2 → count = 4
1     → remove 1 → count = 5
0     → stop
```

**Complexity:**
- Time: O(log₁₀N) - divides by 10 each time
- Space: O(1)

---

## Method 2: Log Method (Optimal)

### Python
```python
import math

def count_digits(n):
    return int(math.log10(n) + 1)

print(count_digits(12345))  # Output: 5
```

### C++
```cpp
#include <cmath>

int countDigits(int n) {
    return (int)log10(n) + 1;
}
```

**How it works:**
```
12345 → log₁₀(12345) = 4.09 → 4 + 1 = 5
100   → log₁₀(100) = 2.0 → 2 + 1 = 3
7     → log₁₀(7) = 0.84 → 0 + 1 = 1
```

**Why +1?** 
- log₁₀(100) = 2, but 100 has **3** digits
- log₁₀(999) = 2.99, but 999 has **3** digits

**Complexity:**
- Time: O(1) - single operation
- Space: O(1)

---

## Quick Comparison

| Method | Time | Space | Code |
|--------|------|-------|------|
| Loop | O(log₁₀N) | O(1) | Manual division |
| Log | **O(1)** | O(1) | `log10(n) + 1` |
| String | O(log₁₀N) | O(log₁₀N) | `len(str(n))` |

# GCD (Greatest Common Divisor) - Python Notes

## Problem
Find the largest number that divides both numbers evenly.

**Example:** GCD(20, 15) = 5

---

## Method 1: Euclidean (Modulo) - **Best**

```python
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

print(gcd(20, 15))  # Output: 5
```

**How it works:**
```
GCD(20, 15)
→ GCD(15, 5)   # 20 % 15 = 5
→ GCD(5, 0)    # 15 % 5 = 0
→ 5
```

**Complexity:** O(log min(a,b))

---

## Method 2: Recursive (Clean)

```python
def gcd(a, b):
    if b == 0:
        return a
    return gcd(b, a % b)

print(gcd(20, 15))  # Output: 5
```

---

## Method 3: Built-in (Easiest)

```python
import math

print(math.gcd(20, 15))  # Output: 5

# Multiple numbers
print(math.gcd(20, 15, 10))  # Output: 5
```

---

## LCM (Bonus)

**Formula:** `LCM(a, b) = (a × b) // GCD(a, b)`

```python
def lcm(a, b):
    return (a * b) // math.gcd(a, b)

print(lcm(20, 15))  # Output: 60
```

Or built-in (Python 3.9+):
```python
import math
print(math.lcm(20, 15))  # Output: 60
```

---

## Quick Tips

✅ Use `math.gcd()` for LeetCode  
✅ Use loop method for interviews  
✅ GCD(a, 0) = a  
✅ GCD(a, 1) = 1
