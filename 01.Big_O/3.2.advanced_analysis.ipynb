```python
# Recursive Function Analysis Notes
# Function 2: Palindrome Check Function
def recursive_function2(mystring, a, b):
   if(a >= b):                                 # 2 # Base case
       return True                             # 1
   else:                                       # Recursive case
       if(mystring[a] != mystring[b]):         # 2
           return False                        # 1
       else:
           return recursive_function2(mystring, a+1, b-1)  # 3 + T(n-2)

def function2(mystring):
   return recursive_function2(mystring, 0, len(mystring)-1)  # 3 + T(n)

'''
T(n) = {
   Base Case:
   If a >= b: T(0) = 3 (comparison + return)
   Recursive Case:
   8 + T(n-2)
}
'''

# Recurrence Relation:
# T(n) = 8 + T(n-2)
# T(n-2) = 8 + T(n-4)
# T(n-4) = 8 + T(n-6)
# ...

# Substitution Method:
# T(n) = 8 + T(n-2)
#      = 8 + 8 + T(n-4)
#      = 8 + 8 + 8 + T(n-6)
#      = ...
#      = T(n) = 8*k + T(n-2k), here loop runs till base case i.e T(0) or T(1)

# Where:
# T(n-2k) = T(0) or T(1)
# n - 2k = 0 or 1
# k = n/2 (approximately)

# Final Solution:
# T(n) = 8*k + T(n-2k) = 8(n/2) + 3
#      = 4n + 3
# T(n) = O(n)
```

```python
# Recursive Function Analysis Notes
# Function 3: Optimized Power Function
def recursive_function3(value, number):
   if (number == 0):                           # 1 # Base case 1
       return 1                                # 1
   elif (number == 1):                         # 1 # Base case 2  
       return value                            # 1
   else:                                       # Recursive case
       half = number // 2                      # 2
       result = recursive_function3(value, half)# 1 + T(n/2)
       if (number % 2 == 0):                   # 2
           return result * result              # 2
       else:
           return value * result * result      # 3

'''
T(n) = {
   Base Cases:
   If n = 0: T(0) = 2 (comparison + return)
   If n = 1: T(1) = 3 (comparison + comparison + return)
   Recursive Case:
   10 + T(n/2)
}
'''

# Recurrence Relation:
# T(n) = 10 + T(n/2)
# T(n/2) = 10 + T(n/4)
# T(n/4) = 10 + T(n/8)
# ...

# Substitution Method:
# T(n) = 10 + T(n/2)
#      = 10 + 10 + T(n/4)
#      = 10 + 10 + 10 + T(n/8)
#      = ...
#      = T(n) = 10*k + T(n/2^k), here loop runs till base case i.e T(1)

# Where:
# T(n/2^k) = T(1)
# n/2^k = 1
# n = 2^k
# k = log₂ n

# Final Solution:
# T(n) = 10*k + T(n/2^k) = 10(log₂ n) + 3
#      = 10 log₂ n + 3
# T(n) = O(log n)
```
