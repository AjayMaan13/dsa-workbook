```python
# Stack Notes

'''
What is a Stack?
- Ordered collection where addition and removal happen at the same end
- The end is called "top", opposite end is "base"
- LIFO principle: Last-In First-Out
- Newest items near top, oldest items near base
'''

# Stack Operations:
# - Push: Add item to top of stack
# - Pop: Remove item from top of stack

'''
Key Properties:
- Items closer to base have been in stack longest
- Most recently added item is first to be removed
- Reverses order of items (insertion order is reverse of removal order)
'''

# Real-world Examples:
# - Web browser Back button (URLs stored in stack)
# - Undo functionality in applications
# - Function call stack in programming
# - Stack of plates in cafeteria

'''
Visual Example:
Push sequence: 1, 2, 3, 4, 5, 6
Stack after pushes: [1, 2, 3, 4, 5, 6] (6 at top)
Pop sequence: 6, 5, 4, 3, 2, 1 (reverse order)
'''

# Time Complexity:
# - Push: O(1)
# - Pop: O(1)
# - Peek/Top: O(1)

'''
Applications:
- Expression evaluation
- Syntax parsing
- Backtracking algorithms
- Memory management
'''
```

```python
'''
The stack abstract data type is defined by the following structure and operations. 
A stack is structured, as described above, as an ordered collection of items where items are added to 
and removed from the end called the “top.” Stacks are ordered LIFO. The stack operations are given below.

- Stack() creates a new stack that is empty. It needs no parameters and returns an empty stack.
- push(item) adds a new item to the top of the stack. It needs the item and returns nothing.
- pop() removes the top item from the stack. It needs no parameters and returns the item. The stack is modified.
- peek() returns the top item from the stack but does not remove it. It needs no parameters. The stack is not modified.
- isEmpty() tests to see whether the stack is empty. It needs no parameters and returns a boolean value.
- size() returns the number of items on the stack. It needs no parameters and returns an integer.
'''
```

```python
class Stack:
    
    
    def __init__(self):
        self.items = []

    def isEmpty(self):
        return self.items == []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        return self.items.pop()

    def peek(self):
        return self.items[len(self.items)-1]

    def size(self):
        return len(self.items)
```

## Stack Using Queue

Implement a stack using two queues (`collections.deque`).

**Two approaches:**
- **Push costly:** On every push, rotate the queue so the new element is at the front (top of stack). Pop is then O(1).
- **Pop costly:** Push is O(1), but pop requires draining q1 into q2 to expose the last element.

We use the **push-costly** approach below:

| Operation | Time Complexity |
|-----------|----------------|
| push      | O(n)           |
| pop       | O(1)           |
| peek      | O(1)           |
| isEmpty   | O(1)           |
| size      | O(1)           |


```python
from collections import deque

class StackUsingQueue:
    """
    Stack implemented with two queues (push-costly approach).

    On each push, the new item is enqueued into q2, then all items
    from q1 are moved into q2, and finally q1 and q2 are swapped.
    This keeps q1 ordered so the top of the stack is always at the
    front of q1 — making pop and peek O(1).
    """

    def __init__(self):
        self.q1 = deque()  # main queue; front = top of stack
        self.q2 = deque()  # helper queue used during push

    def push(self, item):
        # Enqueue new item into the empty helper queue
        self.q2.append(item)
        # Move all existing items behind the new item
        while self.q1:
            self.q2.append(self.q1.popleft())
        # Swap: q1 becomes the updated queue, q2 is emptied
        self.q1, self.q2 = self.q2, self.q1

    def pop(self):
        if self.isEmpty():
            raise IndexError("pop from empty stack")
        return self.q1.popleft()

    def peek(self):
        if self.isEmpty():
            raise IndexError("peek from empty stack")
        return self.q1[0]

    def isEmpty(self):
        return len(self.q1) == 0

    def size(self):
        return len(self.q1)
```

```python
# --- Demo ---
sq = StackUsingQueue()

print(sq.isEmpty())       # True

sq.push(1)
sq.push('two')
sq.push(True)

print(sq.peek())          # True  (last pushed = top)
print(sq.size())          # 3

print(sq.pop())           # True
print(sq.pop())           # 'two'
print(sq.pop())           # 1

print(sq.size())          # 0
print(sq.isEmpty())       # True
```

## Stack Using Queue (List as Queue)

Same push-costly approach, but uses a plain `list` instead of `collections.deque`.

- `append()` → enqueue (add to back)
- `pop(0)` → dequeue (remove from front) — O(n) due to shifting, but functionally identical


```python
class StackUsingList:
    """
    Stack implemented with two plain lists used as queues (push-costly approach).
    list.append()  -> enqueue (O(1))
    list.pop(0)    -> dequeue (O(n) due to shifting, but queue semantics are preserved)
    """

    def __init__(self):
        self.q1 = []  # main queue; front (index 0) = top of stack
        self.q2 = []  # helper queue used during push

    def push(self, item):
        self.q2.append(item)
        while self.q1:
            self.q2.append(self.q1.pop(0))
        self.q1, self.q2 = self.q2, self.q1

    def pop(self):
        if self.isEmpty():
            raise IndexError("pop from empty stack")
        return self.q1.pop(0)

    def peek(self):
        if self.isEmpty():
            raise IndexError("peek from empty stack")
        return self.q1[0]

    def isEmpty(self):
        return len(self.q1) == 0

    def size(self):
        return len(self.q1)
```

```python
# --- Demo ---
sl = StackUsingList()

print(sl.isEmpty())       # True

sl.push(1)
sl.push('two')
sl.push(True)

print(sl.peek())          # True
print(sl.size())          # 3

print(sl.pop())           # True
print(sl.pop())           # 'two'
print(sl.pop())           # 1

print(sl.size())          # 0
print(sl.isEmpty())       # True
```
