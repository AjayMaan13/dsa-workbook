# Infix, Prefix & Postfix Expressions

Mathematical expressions can be written in three notations depending on where the **operator** is placed relative to its **operands**.

| Notation    | Operator position        | Example         |
|-------------|--------------------------|------------------|
| **Infix**   | Between operands         | `A + B`          |
| **Prefix**  | Before operands          | `+ A B`          |
| **Postfix** | After operands           | `A B +`          |

Infix is what humans use. Computers prefer **postfix** (also called Reverse Polish Notation) because it eliminates the need for parentheses and precedence lookups during evaluation.

---
## Operator Precedence & Associativity

When converting or evaluating expressions, operators are processed according to precedence (higher = binds tighter) and associativity (left-to-right or right-to-left).

| Precedence | Operator(s)         | Symbol(s)   | Associativity |
|------------|---------------------|-------------|---------------|
| 4 (highest)| Parentheses         | `( )`       | N/A — override all rules |
| 3          | Exponentiation      | `^` or `**` | Right-to-left |
| 2          | Multiplication / Division | `*` `/` | Left-to-right |
| 1 (lowest) | Addition / Subtraction    | `+` `-` | Left-to-right |

**Key rules:**
- Higher precedence operators are evaluated **before** lower ones.
- Same precedence → use **associativity** to decide order.
- Parentheses **always override** precedence — innermost first.
- Exponentiation is **right-associative**: `2^3^2` = `2^(3^2)` = `2^9` = 512, NOT `(2^3)^2` = 64.

---
## Infix

The standard mathematical notation where the operator sits **between** the operands.

**Pros:** Natural and readable for humans.  
**Cons:** Requires precedence rules and parentheses to resolve ambiguity.

### Easy Examples
| Infix Expression | Meaning                      |
|------------------|------------------------------|
| `A + B`          | Add A and B                  |
| `A - B * C`      | Subtract (B×C) from A        |
| `(A + B) * C`    | Add A+B first, then multiply |

### Complex Examples
| Infix Expression              | Notes                                      |
|-------------------------------|--------------------------------------------|
| `A + B * C - D / E`          | `*` and `/` before `+` and `-`             |
| `(A + B) * (C - D) / E`      | Parentheses group additions/subtractions   |
| `A ^ B ^ C`                  | Right-associative: `A ^ (B ^ C)`           |
| `(A + B) * C ^ D - E / F`    | Mixed precedence with power and brackets   |

---
## Prefix (Polish Notation)

The operator is placed **before** its operands. Introduced by Jan Łukasiewicz.

**Pros:** No parentheses needed; easy for recursive evaluation.  
**Cons:** Less intuitive for humans.

**Evaluation:** Read right-to-left, push operands onto a stack; when an operator is found, pop two operands, apply the operator, push result.

### Easy Examples
| Infix         | Prefix      |
|---------------|-------------|
| `A + B`       | `+ A B`     |
| `A - B`       | `- A B`     |
| `A * B + C`   | `+ * A B C` |
| `(A + B) * C` | `* + A B C` |

### Complex Examples
| Infix                          | Prefix                   |
|--------------------------------|--------------------------|
| `A + B * C - D`                | `- + A * B C D`          |
| `(A + B) * (C - D)`            | `* + A B - C D`          |
| `A ^ B ^ C`                    | `^ A ^ B C`              |
| `(A + B) * C ^ D - E / F`      | `- * + A B ^ C D / E F`  |

---
## Postfix (Reverse Polish Notation — RPN)

The operator is placed **after** its operands.

**Pros:** No parentheses or precedence rules needed during evaluation; stack-based evaluation is simple and fast.  
**Cons:** Less readable for humans.

**Evaluation:** Read left-to-right; push operands onto a stack; when an operator is found, pop two operands, apply the operator, push result.

### Easy Examples
| Infix         | Postfix     |
|---------------|-------------|
| `A + B`       | `A B +`     |
| `A - B`       | `A B -`     |
| `A * B + C`   | `A B * C +` |
| `(A + B) * C` | `A B + C *` |

### Complex Examples
| Infix                          | Postfix                   |
|--------------------------------|---------------------------|
| `A + B * C - D`                | `A B C * + D -`           |
| `(A + B) * (C - D)`            | `A B + C D - *`           |
| `A ^ B ^ C`                    | `A B C ^ ^`               |
| `(A + B) * C ^ D - E / F`      | `A B + C D ^ * E F / -`   |

---
## Quick Conversion Reference

```
Infix:   (A + B) * C ^ D - E / F
Prefix:  - * + A B ^ C D / E F
Postfix: A B + C D ^ * E F / -
```

**Infix → Postfix (Shunting-Yard algorithm sketch):**
1. Read tokens left-to-right.
2. Operand → output immediately.
3. Operator → pop stack to output while stack-top has **higher or equal** precedence (respecting associativity), then push current operator.
4. `(` → push onto stack.
5. `)` → pop to output until `(` is found; discard both parentheses.
6. End → pop all remaining operators to output.

**Infix → Prefix:**  
Reverse the infix expression (swap `(` ↔ `)`), apply Shunting-Yard, then reverse the output.
