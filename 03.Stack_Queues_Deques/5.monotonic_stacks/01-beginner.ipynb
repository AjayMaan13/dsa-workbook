# Monotonic Stacks — Beginner

---

## What even IS a monotonic stack?

It's just a **regular stack** with one rule: elements must always be in a specific order.

- **Monotonic Increasing** → each new element is ≥ the one below it
- **Monotonic Decreasing** → each new element is ≤ the one below it

That's it. The magic is in *what happens when you break the rule* — you **pop** elements until the rule is satisfied again.

---

## The Core Intuition

> A monotonic stack is useful when you need to find, for each element, the **next/previous greater or smaller element**.

These are called **"Next Greater Element"** type problems — and they're the backbone of almost every monotonic stack question.

## Quick Stack Recap

Before anything, let's remind ourselves how a stack works.

```python
stack = []
stack.append(3)   # push 3  →  [3]
stack.append(7)   # push 7  →  [3, 7]
stack.pop()       # pop  7  →  [3]
print(stack[-1])  # peek    →  3  (top of stack, no removal)
```

---

## Walkthrough: Next Greater Element

**Problem:** Given `[2, 1, 5, 3, 7]`, find the next greater element for each number.

**Expected output:** `[5, 5, 7, 7, -1]`

**Strategy:** Use a **monotonic decreasing stack** (stores indices of elements in decreasing order).
When a new element breaks the rule (is bigger), that element IS the "next greater" for whatever we pop.

### Step-by-step trace:

| Step | Element | Stack (values) | result[] |
|------|---------|----------------|----------|
| i=0  | 2       | Push. → `[2]`            | `[?, ?, ?, ?, ?]` |
| i=1  | 1       | 1 < 2, no pop. Push. → `[2, 1]` | `[?, ?, ?, ?, ?]` |
| i=2  | 5       | 5 > 1 → pop, result[1]=5. 5 > 2 → pop, result[0]=5. Push. → `[5]` | `[5, 5, ?, ?, ?]` |
| i=3  | 3       | 3 < 5, no pop. Push. → `[5, 3]` | `[5, 5, ?, ?, ?]` |
| i=4  | 7       | 7 > 3 → pop, result[3]=7. 7 > 5 → pop, result[2]=7. Push. → `[7]` | `[5, 5, 7, 7, ?]` |
| End  | —       | idx 4 left → result[4] stays -1 | `[5, 5, 7, 7, -1]` ✓ |

> **Key insight:** Every pop is an answer being filled in. When you pop X because new element Y is greater, Y is the Next Greater Element for X.

---

## The Code

```python
def next_greater_element(nums):
    n = len(nums)
    result = [-1] * n   # default: no next greater element
    stack = []          # stores INDICES (not values!)

    for i in range(n):
        # while current element breaks the "decreasing" rule
        while stack and nums[stack[-1]] < nums[i]:
            idx = stack.pop()
            result[idx] = nums[i]   # nums[i] is the NGE for idx
        stack.append(i)

    return result


nums = [2, 1, 5, 3, 7]
print(next_greater_element(nums))   # [5, 5, 7, 7, -1]
```

### Why store indices, not values?

You need to know **where** to write the answer (`result[idx] = ...`).  
If you store values, you lose the position.  
Always push indices — get the value back with `nums[stack[-1]]` when needed.

## Visualising the Monotonic Property

Let's print each step so you can see the stack evolving in real time.

```python
def next_greater_verbose(nums):
    n = len(nums)
    result = [-1] * n
    stack = []

    for i in range(n):
        print(f"\n--- i={i}, val={nums[i]} ---")
        while stack and nums[stack[-1]] < nums[i]:
            idx = stack.pop()
            result[idx] = nums[i]
            print(f"  Popped idx={idx} (val={nums[idx]}), NGE = {nums[i]}")
        stack.append(i)
        print(f"  Stack (values): {[nums[j] for j in stack]}")
        print(f"  Result so far:  {result}")

    return result


next_greater_verbose([2, 1, 5, 3, 7])
```

## The Pattern You'll Recognise Every Time

Whenever you see these phrases, think **monotonic stack**:

- "next greater element"
- "next smaller element"
- "how many days until a warmer temperature"
- "largest rectangle in histogram"
- "stock span problem"

The template is always:
1. Iterate through the array
2. While the stack isn't empty AND current element breaks your rule → **pop and record**
3. Push current index onto the stack

| Stack type | Pop when current is... | Finds... |
|------------|------------------------|----------|
| Decreasing | bigger                 | Next Greater Element |
| Increasing | smaller                | Next Smaller Element |

## Practice: Daily Temperatures (LeetCode 739)

Given daily temperatures, return how many days until a warmer day (0 if none).

`temps = [73, 74, 75, 71, 69, 72, 76, 73]`  
Expected: `[1, 1, 4, 2, 1, 1, 0, 0]`

It's the exact same template — the only change is `result[idx] = i - idx` (distance) instead of the value.

```python
def daily_temperatures(temps):
    n = len(temps)
    result = [0] * n
    stack = []   # indices

    for i in range(n):
        while stack and temps[stack[-1]] < temps[i]:
            idx = stack.pop()
            result[idx] = i - idx   # <-- only change: store distance
        stack.append(i)

    return result


temps = [73, 74, 75, 71, 69, 72, 76, 73]
print(daily_temperatures(temps))   # [1, 1, 4, 2, 1, 1, 0, 0]
```

## Common Beginner Mistakes

| Mistake | Fix |
|---------|-----|
| Storing **values** instead of **indices** | Always push `i`, retrieve value as `nums[i]` |
| Forgetting leftover elements in the stack | Their answer stays `-1` (or `0`) — make sure your default is correct |
| Confusing next vs previous direction | Direction of iteration controls this — next = left-to-right |
| Using `<=` vs `<` on duplicates | Trace `[3, 3, 3]` manually to verify your comparison |
