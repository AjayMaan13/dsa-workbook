# Longest Subarray with Sum K - Notes

## Problem
Find the length of the longest subarray with sum equal to K.

**Example:** `nums = [10, 5, 2, 7, 1, 9]`, `k = 15`  
**Output:** `4` (subarray: `[2, 7, 1, 5]` or `[5, 2, 7, 1]`)

---

## Sliding Window Approach

```python
def longestSubarray(nums, k):
    n = len(nums)
    maxLen = 0
    left = 0
    right = 0
    sum = nums[0]
    
    while right < n:
        # Shrink window if sum > k
        while left <= right and sum > k:
            sum -= nums[left]
            left += 1
        
        # Update max length if sum == k
        if sum == k:
            maxLen = max(maxLen, right - left + 1)
        
        # Expand window
        right += 1
        if right < n:
            sum += nums[right]
    
    return maxLen
```

**Time:** O(N)  
**Space:** O(1)

---

## How It Works

### Example: `[10, 5, 2, 7, 1, 9]`, k = 15

```
Initial: left=0, right=0, sum=10

Step 1: sum=10 < 15
  right=1, sum=10+5=15 ✅
  maxLen=2 (subarray: [10,5])

Step 2: sum=15 == 15 ✅
  right=2, sum=15+2=17 > 15
  Shrink: sum=17-10=7, left=1

Step 3: sum=7 < 15
  right=3, sum=7+7=14 < 15

Step 4: sum=14 < 15
  right=4, sum=14+1=15 ✅
  maxLen=4 (subarray: [5,2,7,1])

Continue...

Final: maxLen=4
```

---

## Algorithm Steps

1. **Initialize:** `left=0, right=0, sum=nums[0]`
2. **Expand:** Move `right`, add `nums[right]` to `sum`
3. **Shrink:** If `sum > k`, move `left`, subtract `nums[left]`
4. **Check:** If `sum == k`, update `maxLen`
5. **Repeat** until `right` reaches end

---

## Key Points

✅ **Two pointers** - `left` and `right` form sliding window  
✅ **Expand window** - move `right` to add elements  
✅ **Shrink window** - move `left` when sum exceeds k  
✅ **Track max** - update when sum equals k  
✅ **O(N) time** - each element visited at most twice  

---

## Complexity

**Time:** O(N)  
- Each element added once (right pointer)
- Each element removed at most once (left pointer)

**Space:** O(1)  
- Only uses variables: left, right, sum, maxLen

---

## When to Use

✅ Array problems with contiguous subarrays  
✅ Finding subarray with specific sum/property  
✅ Positive integers only (for this approach)  

**Note:** For arrays with negatives, use HashMap approach instead!
