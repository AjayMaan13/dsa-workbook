# Longest Subarray with Sum K - Notes

## Problem
Find the length of the longest subarray with sum equal to K.

**Example:** `nums = [10, 5, 2, 7, 1, 9]`, `k = 15`  
**Output:** `4` (subarray: `[2, 7, 1, 5]` or `[5, 2, 7, 1]`)

---

## Sliding Window Approach

```python
def longestSubarray(nums, k):
    n = len(nums)
    maxLen = 0
    left = 0
    right = 0
    sum = nums[0]
    
    while right < n:
        # Shrink window if sum > k
        while left <= right and sum > k:
            sum -= nums[left]
            left += 1
        
        # Update max length if sum == k
        if sum == k:
            maxLen = max(maxLen, right - left + 1)
        
        # Expand window
        right += 1
        if right < n:
            sum += nums[right]
    
    return maxLen
```

**Time:** O(N)  
**Space:** O(1)

---

## How It Works

### Example: `[10, 5, 2, 7, 1, 9]`, k = 15

```
Initial: left=0, right=0, sum=10

Step 1: sum=10 < 15
  right=1, sum=10+5=15 ✅
  maxLen=2 (subarray: [10,5])

Step 2: sum=15 == 15 ✅
  right=2, sum=15+2=17 > 15
  Shrink: sum=17-10=7, left=1

Step 3: sum=7 < 15
  right=3, sum=7+7=14 < 15

Step 4: sum=14 < 15
  right=4, sum=14+1=15 ✅
  maxLen=4 (subarray: [5,2,7,1])

Continue...

Final: maxLen=4
```

## Visual Trace

---

## Algorithm Steps

1. **Initialize:** `left=0, right=0, sum=nums[0]`
2. **Expand:** Move `right`, add `nums[right]` to `sum`
3. **Shrink:** If `sum > k`, move `left`, subtract `nums[left]`
4. **Check:** If `sum == k`, update `maxLen`
5. **Repeat** until `right` reaches end

---

## Key Points

✅ **Two pointers** - `left` and `right` form sliding window  
✅ **Expand window** - move `right` to add elements  
✅ **Shrink window** - move `left` when sum exceeds k  
✅ **Track max** - update when sum equals k  
✅ **O(N) time** - each element visited at most twice  

---

## Complexity

**Time:** O(N)  
- Each element added once (right pointer)
- Each element removed at most once (left pointer)

**Space:** O(1)  
- Only uses variables: left, right, sum, maxLen

---

## When to Use

✅ Array problems with contiguous subarrays  
✅ Finding subarray with specific sum/property  
✅ Positive integers only (for this approach)  

**Note:** For arrays with negatives, use HashMap approach instead!

# Longest Subarray with Sum K (Positives & Negatives)

**Problem:** Given an array containing positive and negative integers, find the length of the longest subarray that sums to k.

---

## Optimal Approach - Prefix Sum + HashMap

### Core Concept
**Prefix Sum Logic:**
- If prefix sum at index `j` is `sum_j` and at index `i` is `sum_i` (where `i < j`)
- Then sum of subarray from `(i+1)` to `j` = `sum_j - sum_i`
- If `sum_j - sum_i = k`, then `sum_i = sum_j - k`
- **Key Insight:** If we've seen `(current_sum - k)` before, we found a valid subarray!

### Why HashMap?
- Store **first occurrence** of each prefix sum to maximize subarray length
- Quick O(1) lookup to check if `(current_sum - k)` exists

---

## Algorithm Steps

1. **Initialize:**
   - `pre_sum_map = {}` → stores {prefix_sum: first_index}
   - `sum_so_far = 0` → running prefix sum
   - `max_len = 0` → tracks longest subarray length

2. **For each element at index i:**
   - Update running sum
   - **Check two cases:**
     - **Case 1:** Entire array from start has sum = k
     - **Case 2:** A previous prefix sum exists where `(current_sum - that_sum) = k`
   - **Store prefix sum** (only if not already stored)

---

## Implementation with Detailed Comments

```python
class SubarraySolver:
    def get_longest_subarray(self, a, k):
        """
        Find length of longest subarray with sum = k
        Works with positive and negative numbers
        
        Args:
            a: List of integers (can include negatives)
            k: Target sum
            
        Returns:
            Length of longest subarray with sum k
        """
        n = len(a)
        pre_sum_map = {}  # {prefix_sum: first_index_where_it_occurred}
        sum_so_far = 0    # Running prefix sum
        max_len = 0       # Maximum length found so far
        
        for i in range(n):
            # Add current element to running sum
            sum_so_far += a[i]
            
            # CASE 1: Check if subarray from index 0 to i has sum = k
            # This means the entire subarray from start sums to k
            if sum_so_far == k:
                max_len = i + 1  # Length is (i + 1) since we're 0-indexed
            
            # CASE 2: Check if there's a previous prefix sum such that
            # (current_sum - previous_sum) = k
            # This means: previous_sum = current_sum - k
            rem = sum_so_far - k
            
            if rem in pre_sum_map:
                # Found a subarray! Calculate its length
                # Length = current_index - previous_index
                length = i - pre_sum_map[rem]
                max_len = max(max_len, length)
            
            # Store the FIRST occurrence of this prefix sum
            # Why only first? Because we want maximum length!
            # Earlier index = longer potential subarray
            if sum_so_far not in pre_sum_map:
                pre_sum_map[sum_so_far] = i
        
        return max_len


# Example usage with step-by-step trace
def main():
    solver = SubarraySolver()
    
    # Example 1: Array with negatives
    a = [-1, 1, 1]
    k = 1
    length = solver.get_longest_subarray(a, k)
    print(f"Array: {a}, k: {k}")
    print(f"Longest subarray length: {length}")  # Output: 3
    
    # Example 2: Another test case
    a = [1, -1, 5, -2, 3]
    k = 3
    length = solver.get_longest_subarray(a, k)
    print(f"\nArray: {a}, k: {k}")
    print(f"Longest subarray length: {length}")  # Output: 4 ([1,-1,5,-2])


if __name__ == "__main__":
    main()
```

---

## Complexity Analysis

- **Time Complexity:** O(n) - Single pass through array with O(1) hashmap operations
- **Space Complexity:** O(n) - HashMap can store up to n prefix sums in worst case

---

## Why Store Only First Occurrence?

**Example:** `a = [1, -1, 2], k = 2`

| Index | Element | sum_so_far | Action |
|-------|---------|------------|--------|
| 0 | 1 | 1 | Store map[1] = 0 |
| 1 | -1 | 0 | Store map[0] = 1 |
| 2 | 2 | 2 | sum == k, max_len = 3 |

If we stored **last occurrence** of sum=0 at index 1, we'd miss the longer subarray starting from index 0!

---

## Key Differences from "Positives Only" Version

| Aspect | Positives Only | With Negatives |
|--------|---------------|----------------|
| Approach | Sliding Window (Two Pointer) | Prefix Sum + HashMap |
| Why? | Sum always increases | Sum can increase/decrease |
| Time | O(n) | O(n) |
| Space | O(1) | O(n) |
