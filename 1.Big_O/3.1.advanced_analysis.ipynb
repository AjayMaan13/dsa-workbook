```python
# Algorithm Complexity Analysis Notes

## Function 2: Mathematical Expression Analysis

def function1(list):
    n = len(list)                           # 1
    for i in range(n - 1):                  # 2 + n - 1
        for j in range(0, len(list) - 1 - i):   # 4 + S₁
            if list[j] > list[j+1]:         # 2 * (S₁), considering list[] as 0 operation
                tmp = list[j]               # 1 * (S₁)
                list[j] = list[j+1]         # 2 * (S₁)
                list[j+1] = tmp             # 2 * (S₁)

### Operation Count Table
'''
# Operation Count Table for Function3 (Bubble Sort)
# 
# i     | Inner Loop Iterations | Ops per Iteration = 4 + (n - 1 - i)
# ------|----------------------|-------------------------------------
# 0     | n - 1               | 4 + (n - 1) = n + 3
# 1     | n - 2               | 4 + (n - 2) = n + 2  
# 2     | n - 3               | 4 + (n - 3) = n + 1
# ...   | ...                 | ...
# n-2   | 1                   | 4 + 1 = 5

S₁ = 5 + 6 + 7 + ... + (n + 1) + (n + 2) + (n + 3)
S₁ + 10 = 1 + 2 + 3 + 4 + 5 + 6 + 7 + ... + (n + 1) + (n + 2) + (n + 3)
S₁ + 10 = (n + 3)(n + 4)/2
S₁ = (n + 3)(n + 4)/2 - 10
S₁ = (n² + 7n + 12)/2 - 10
S₁ = (n² + 7n + 12 - 20)/2
S₁ = (n² + 7n - 8)/2
S₁ = 0.5n² + 3.5n - 4
'''

# T(n): 1 + (2 + n - 1) + (4 + S₁) + (2 * S₁) + (1 * S₁) + (2 * S₁) + (2 * S₁)
# T(n): 1 + (2 + n - 1) + (4 + S₁) + (7 * S₁)
# T(n): 6 + n + 8S₁

# Where S₁ = 0.5n² + 3.5n - 4

# Substituting S₁:
# T(n): 6 + n + 8(0.5n² + 3.5n - 4)
# T(n): 6 + n + 4n² + 28n - 32
# T(n): 4n² + 29n - 26

# Therefore: T(n) = 4n² + 29n - 26 = O(n²)
```

```python
def function2(mylist, key):
   total = 0                                   # 1
   for i in range(len(mylist)):                # 2 + n
       for j in range(i+1, len(mylist)):       # 3 + S₁
           if i != j:                          # 1 * S₁
               if mylist[i] + mylist[j] == key: # 2 * S₁, considering list[] as 0 operation
                   total += 1                  # 2 * S₁
   return total                                # 1


# Operation Count Table for Inner Loop Iterations
#
# i     | Inner Loop Iterations | Ops per Iteration = 3 + (n - 1 - i)
# ------|----------------------|-------------------------------------
# 0     | n - 1               | 3 + (n - 1) = n + 2
# 1     | n - 2               | 3 + (n - 2) = n + 1  
# 2     | n - 3               | 3 + (n - 3) = n
# ...   | ...                 | ...
# n-2   | 1                   | 3 + 1 = 4
# n-1   | 0                   | 3 + 0 = 3

# Mathematical Summation
# S₁ = 3 + 4 + 5 + ... + n + (n + 1) + (n + 2)
# 
# S₁ + 6 = 1 + 2 + 3 + 4 + 5 + ... + n + (n + 1) + (n + 2)
# 
# S₁ + 6 = (n + 2)(n + 3)/2
# 
# S₁ = (n + 2)(n + 3)/2 - 6
# 
# S₁ = (n² + 5n + 6)/2 - 6
# 
# S₁ = (n² + 5n + 6 - 12)/2
# 
# S₁ = (n² + 5n - 6)/2
# 
# S₁ = 0.5n² + 2.5n - 3

# Total: 1 + (2 + n) + (3 + S₁) + (1 * S₁) + (2 * S₁) + (2 * S₁) + 1
# Total: 7 + n + S₁ + 5S₁
# Total: 7 + n + 6S₁

# Substituting S₁:
# Total: 7 + n + 6(0.5n² + 2.5n - 3)
# Total: 7 + n + 3n² + 15n - 18
# Total: 3n² + 16n - 11

# Therefore: T(n) = 3n² + 16n - 11 = O(n²)
```

```python
def function3(mylist, key):
   total = 0                           # 1
   mylist.sort()                       # n log n
   i = 0                              # 1
   j = len(mylist) - 1                # 2
   while (i < j):                     # 1 + S₁
       if(mylist[i] + mylist[j] < key): # 2 * S₁
           i += 1                     # 0 (not executed in worst case)
       elif(mylist[i] + mylist[j] > key): # 2 * S₁
           j -= 1                     # 0 (not executed in worst case)
       else:                          # 1 * S₁
           total += 1                 # 2 * S₁ (worst case: always executed)
           i += 1                     # 2 * S₁ (worst case: always executed)
           j -= 1                     # 2 * S₁ (worst case: always executed)
   return total                       # 1

# Operation Count Table for While Loop (worst case - always else)
#
# iterations | number of elements
# -----------|-------------------
# 1          | n
# 2          | n - 2
# 3          | n - 4
# ...        | ...
# n/2        | 2

# S₁ = number of while loop iterations ≈ n/2 (worst case)
# Each iteration: 12 operations

# Total operations:
# Total: 1 + n log n + 1 + 2 + (1 + S₁) + (2 * S₁) + (2 * S₁) + (1 * S₁) + (2 * S₁) + (2 * S₁) + (2 * S₁) + 1
# Total: 6 + n log n + 12S₁

# Substituting S₁ ≈ n/2:
# Total: 6 + n log n + 12(n/2)
# Total: 6 + n log n + 6n
# Total: n log n + 6n + 6

# Therefore: T(n) = n log n + 6n + 6 = O(n log n)
```
