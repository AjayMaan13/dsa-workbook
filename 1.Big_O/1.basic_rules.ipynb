```python
# What is Big O Notation?
# ------------------------
# Big O notation helps us understand how the time or steps needed by a program
# increase when the input (like a list) gets bigger.
# It’s a way to measure how efficient an algorithm is.

# Imagine you have to do a task with a list of names.
# If you have 10 names or 1000 names, how much longer will it take?
# Big O tells us this!

# ------------------------
# O(1) - Constant time
# ------------------------
# The time it takes does NOT change when the input gets bigger.
# Example: Getting the first item of a list.
def get_first_item(items):
    return items[0]  # Always just one step, no matter the list size

# Think of it like grabbing the first book on a shelf — always quick!

# ------------------------
# O(log n) - Logarithmic time
# ------------------------
# The time grows slowly as input size grows.
# Algorithms like binary search cut the problem in half at each step.
# Example: Searching a sorted list by repeatedly cutting it in half.
def binary_search(sorted_list, target):
    left, right = 0, len(sorted_list) - 1
    while left <= right:
        mid = (left + right) // 2
        if sorted_list[mid] == target:
            return mid
        elif sorted_list[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

# Imagine looking for a word in a dictionary:
# You open in the middle, decide if you go left or right,
# then open in the middle of that half, and so on.

# ------------------------
# O(n) - Linear time
# ------------------------
# Time increases directly with the size of the input.
# Example: Looking at every item in a list once.
def print_all_items(items):
    for item in items:
        print(item)  # The more items, the longer it takes

# It’s like checking every book on a shelf one by one.

# ------------------------
# O(n log n) - Log-linear time
# ------------------------
# Slightly slower than linear but faster than quadratic.
# Usually happens in efficient sorting algorithms.
# Example: Sorting a list (like mergesort or quicksort).

# Imagine sorting cards by repeatedly splitting the deck and merging.

# ------------------------
# O(n^2) - Quadratic time
# ------------------------
# Time grows with the square of the input size.
# If input doubles, time quadruples.
# Example: Checking every pair in a list.
def print_all_pairs(items):
    for item1 in items:
        for item2 in items:
            print(item1, item2)  # Two loops inside each other

# Like shaking hands with everyone at a party:
# If there are 10 people, there are 100 handshakes.

# ------------------------
# O(2^n) - Exponential time
# ------------------------
# Time doubles every time the input increases by 1.
# Example: Simple recursive Fibonacci without optimization.
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

# Very slow for large n, because it repeats many calculations.

# ------------------------
# O(n!) - Factorial time
# ------------------------
# Time grows extremely fast, much worse than exponential.
# Example: Trying every possible order of items.
# Like arranging 5 books in every possible way — a huge number of possibilities!

# ------------------------
# Summary:
# - Big O describes how the time needed grows when input grows.
# - Lower Big O means the algorithm stays faster for big inputs.
# - We ignore small details and constants, focusing on the biggest impact.
# - Understanding Big O helps you write code that works well even with lots of data.

# Hope this helps you understand how algorithms behave as your data grows!
```

```python
# 📚 Big-O Notation Cheat Sheet (Python Notes Style)

# 1️⃣ Big O (O) — Worst-case scenario
# -------------------------------------------------
# Think of this as the maximum time it might take.
# Example: "Even if I'm slow, it won't take longer than this."

# 🧠 Imagine: "I can clean my room in no more than 30 minutes."
# Even if you're lazy, it won’t go past that.

# 📦 In code:
# If a program runs in at most n² steps, we say it’s O(n²)

# 🗣️ Say: "Big O tells us the worst-case time complexity."


# 2️⃣ Little o (o) — Strictly better than that
# -------------------------------------------------
# It means: "I'm definitely faster than that — not even close!"

# 🧠 Imagine: "I always finish cleaning in less than 30 minutes.
# Even if I'm slow, I never reach that mark."

# 📦 In code:
# If your algorithm takes n steps, that’s o(n log n),
# because n log n is always eventually bigger than n.

# 🗣️ Say: "Little o means I’m better than that limit, always."


# 3️⃣ Omega (Ω) — Best-case scenario
# -------------------------------------------------
# This tells us the minimum time something will take.

# 🧠 Imagine: "I need at least 10 minutes to clean my room, 
# even if I speed-run it."

# 📦 In code:
# If something always takes n steps or more, it's Ω(n)

# 🗣️ Say: "Omega tells us the best-case time complexity."


# 4️⃣ Theta (Θ) — Exactly this fast (tight bound)
# -------------------------------------------------
# It means both the best and worst time are the same.

# 🧠 Imagine: "I always take exactly 15 minutes to clean my room —
# not faster, not slower."

# 📦 In code:
# If your program always takes n log n steps, and never more or less,
# we say it’s Θ(n log n)

# 🗣️ Say: "Theta means exact speed — no surprises."


# ⚡ Final Summary (Like a Superhero Team!)
# -------------------------------------------------
# Notation | Meaning           | Like Saying...
# ---------|-------------------|-----------------------------
# O        | Worst-case        | "I won’t be slower than this."
# o        | Strictly faster   | "I’m faster than that."
# Ω        | Best-case         | "I can’t be faster than this."
# Θ        | Exact time        | "This is how fast I always am."
```
